```r
#libraries
library(ggplot2)
library(dplyr) 
library(tidyverse)
library(stringr)
```

```r
# load data
essay_1bar_6cm_1 = read.table("./data/6cm_1bar_1.csv",sep=";",dec=",",header=FALSE)
essay_1bar_6cm_2 = read.table("./data/6cm_1bar_2.csv",sep=";",dec=",",header=FALSE)
essay_1bar_6cm_3 = read.table("./data/6cm_1bar_3.csv",sep=";",dec=",",header=FALSE)
essay_1bar_6cm_4 = read.table("./data/6cm_1bar_4.csv",sep=";",dec=",",header=FALSE)
essay_1bar_8cm_2 = read.table("./data/8cm_1bar_2.csv",sep=";",dec=",",header=FALSE)
essay_1bar_8cm_3 = read.table("./data/8cm_1bar_3.csv",sep=";",dec=",",header=FALSE)
```

```r
#change column names
colnames(essay_1bar_6cm_1) = c("time","timestamp","pressure_bar")
colnames(essay_1bar_6cm_2) = c("time","timestamp","pressure_bar")
colnames(essay_1bar_6cm_3) = c("time","timestamp","pressure_bar")
colnames(essay_1bar_6cm_4) = c("time","timestamp","pressure_bar")
colnames(essay_1bar_8cm_2) = c("time","timestamp","pressure_bar")
colnames(essay_1bar_8cm_3) = c("time","timestamp","pressure_bar")
```

```r
idx_essay_1bar_6cm_1 = which(essay_1bar_6cm_1$pressure_bar == max(essay_1bar_6cm_1$pressure_bar))
idx_essay_1bar_6cm_2 = which(essay_1bar_6cm_2$pressure_bar == max(essay_1bar_6cm_2$pressure_bar))
idx_essay_1bar_6cm_3 = which(essay_1bar_6cm_3$pressure_bar == max(essay_1bar_6cm_3$pressure_bar))
idx_essay_1bar_6cm_4 = which(essay_1bar_6cm_4$pressure_bar == max(essay_1bar_6cm_4$pressure_bar))
idx_essay_1bar_8cm_2 = which(essay_1bar_8cm_2$pressure_bar == max(essay_1bar_8cm_2$pressure_bar))
idx_essay_1bar_8cm_3 = which(essay_1bar_8cm_3$pressure_bar == max(essay_1bar_8cm_3$pressure_bar))
```

```r
subset_essay_1bar_6cm_1 = essay_1bar_6cm_1[(idx_essay_1bar_6cm_1-20):(idx_essay_1bar_6cm_1+10),]
subset_essay_1bar_6cm_2 = essay_1bar_6cm_2[(idx_essay_1bar_6cm_2-20):(idx_essay_1bar_6cm_2+10),]
subset_essay_1bar_6cm_3 = essay_1bar_6cm_3[(idx_essay_1bar_6cm_3-20):(idx_essay_1bar_6cm_3+10),]
subset_essay_1bar_6cm_4 = essay_1bar_6cm_4[(idx_essay_1bar_6cm_4-20):(idx_essay_1bar_6cm_4+10),]
#subset_essay_1bar_8cm_2 = essay_1bar_8cm_2[
subset_essay_1bar_8cm_3 = essay_1bar_8cm_3[(idx_essay_1bar_8cm_3-20):(idx_essay_1bar_8cm_3+10),]
```

```r
#spaghetti plot 
ggplot() +
geom_point(data=subset_essay_1bar_6cm_1,aes(x=seq(0,30,1),y=pressure_bar,colour="6 cm 1 bar")) + 
geom_line(data=subset_essay_1bar_6cm_1,aes(x=seq(0,30,1),y=pressure_bar,colour="6 cm 1 bar")) + 

geom_point(data=subset_essay_1bar_6cm_2,aes(x=seq(0,30,1),y=pressure_bar,colour="6 cm 1 bar")) + 
geom_line(data=subset_essay_1bar_6cm_2,aes(x=seq(0,30,1),y=pressure_bar,colour="6 cm 1 bar")) + 

geom_point(data=subset_essay_1bar_6cm_3,aes(x=seq(0,30,1),y=pressure_bar,colour="6 cm 1 bar")) + 
geom_line(data=subset_essay_1bar_6cm_3,aes(x=seq(0,30,1),y=pressure_bar,colour="6 cm 1 bar")) + 

geom_point(data=subset_essay_1bar_6cm_4,aes(x=seq(0,30,1),y=pressure_bar,colour="6 cm 1 bar")) + 
geom_line(data=subset_essay_1bar_6cm_4,aes(x=seq(0,30,1),y=pressure_bar,colour="6 cm 1 bar")) + 

#geom_point(data=subset_essay_1bar_8cm_2,aes(x=seq(0,30,1),y=pressure_bar,colour="8 cm 1 bar")) + 
#geom_line(data=subset_essay_1bar_8cm_2,aes(x=seq(0,30,1),y=pressure_bar,colour="8 cm 1 bar")) + 

geom_point(data=subset_essay_1bar_8cm_3,aes(x=seq(0,30,1),y=pressure_bar,colour="8 cm 1 bar")) + 
geom_line(data=subset_essay_1bar_8cm_3,aes(x=seq(0,30,1),y=pressure_bar,colour="8 cm 1 bar")) + 

scale_y_continuous(breaks=seq(from = 0, to = 7, by = 0.5)) +
scale_x_continuous(breaks=seq(from = 0, to = 30, by = 5)) +
theme_bw(base_size=12) + 
labs(colour="Legend",x="Time (seconds)", y="Pressure(bar)",title="Burst Essay Results in PVA tubes \n (5 mm internal diameter)")+
theme(panel.grid.major.x=element_line(colour="gray90",size=0.3),
     panel.grid.major.y=element_line(colour="gray90",size=0.3),
     panel.grid.minor.x=element_line(colour="gray90",size=0.3),
      panel.grid.minor.y=element_line(colour="gray90",size=0.3))
```
