```r
# load data
essay_0.1_4_0 = read.csv("./data/0.1-4-0.csv",header=FALSE)
essay_0.1_4_1 = read.csv("./data/0.1-4-1.csv",header=FALSE)
essay_0.1_4_2 = read.csv("./data/0.1-4-2.csv",header=FALSE)
essay_0.1_4_3 = read.csv("./data/0.1-4-3.csv",header=FALSE)
essay_0.1_4_4 = read.csv("./data/0.1-4-4.csv",header=FALSE)
essay_0.1_4_5 = read.csv("./data/0.1-4-5.csv",header=FALSE)
essay_0.3_4_0 = read.csv("./data/0.1-4-6.csv",header=FALSE)
essay_0.1_5_0 = read.csv("./data/0.1-5-0.csv",header=FALSE)
essay_0.1_5_1 = read.csv("./data/0.1-5-1.csv",header=FALSE)
essay_0.3_5_0 = read.csv("./data/0.3-5-0.csv",header=FALSE)
essay_0.3_5_1 = read.csv("./data/0.3-5-1.csv",header=FALSE)
```

```r
#change column names
colnames(essay_0.1_4_0) = c("time","timestamp","pressure_bar")
colnames(essay_0.1_4_1) = c("time","timestamp","pressure_bar")
colnames(essay_0.1_4_2) = c("time","timestamp","pressure_bar")
colnames(essay_0.1_4_3) = c("time","timestamp","pressure_bar")
colnames(essay_0.1_4_4) = c("time","timestamp","pressure_bar")
colnames(essay_0.1_4_5) = c("time","timestamp","pressure_bar")
colnames(essay_0.3_4_0) = c("time","timestamp","pressure_bar")
colnames(essay_0.1_5_0) = c("time","timestamp","pressure_bar")
colnames(essay_0.1_5_1) = c("time","timestamp","pressure_bar")
colnames(essay_0.3_5_0) = c("time","timestamp","pressure_bar")
colnames(essay_0.3_5_1) = c("time","timestamp","pressure_bar")
```

```r
#spaghetti plot 
ggplot() +
geom_point(data=essay_0.1_4_0,aes(x=timestamp,y=pressure_bar,colour="4 cm 0.1 bar")) + 
geom_line(data=essay_0.1_4_0,aes(x=timestamp,y=pressure_bar,colour="4 cm 0.1 bar")) + 

#geom_point(data=essay_0.1_4_2,aes(x=timestamp,y=pressure_bar,colour="4 cm 0.1 bar")) + 
#geom_line(data=essay_0.1_4_2,aes(x=timestamp,y=pressure_bar,colour="4 cm 0.1 bar")) +

#geom_point(data=essay_0.1_4_3,aes(x=timestamp,y=pressure_bar,colour="4 cm 0.1 bar")) + 
#geom_line(data=essay_0.1_4_3,aes(x=timestamp,y=pressure_bar,colour="4 cm 0.1 bar")) + 

#geom_point(data=essay_0.1_4_4,aes(x=timestamp,y=pressure_bar,colour="4 cm 0.1 bar")) + 
#geom_line(data=essay_0.1_4_4,aes(x=timestamp,y=pressure_bar,colour="4 cm 0.1 bar")) +

#geom_point(data=essay_0.1_4_5,aes(x=timestamp,y=pressure_bar,colour="4 cm 0.1 bar")) + 
#geom_line(data=essay_0.1_4_5,aes(x=timestamp,y=pressure_bar,colour="4 cm 0.1 bar"))+ 

geom_point(data=essay_0.3_4_0,aes(x=timestamp,y=pressure_bar,colour="4 cm 0.4 bar")) + 
geom_line(data=essay_0.3_4_0,aes(x=timestamp,y=pressure_bar,colour="4 cm 0.4 bar")) +

geom_point(data=essay_0.1_5_0,aes(x=timestamp,y=pressure_bar,colour="5 cm 0.1 bar")) + 
geom_line(data=essay_0.1_5_0,aes(x=timestamp,y=pressure_bar,colour="5 cm 0.1 bar")) +

geom_point(data=essay_0.1_5_1,aes(x=timestamp,y=pressure_bar,colour="5 cm 0.1 bar")) + 
geom_line(data=essay_0.1_5_1,aes(x=timestamp,y=pressure_bar,colour="5 cm 0.1 bar")) +

geom_point(data=essay_0.3_5_0,aes(x=timestamp,y=pressure_bar,colour="5 cm 0.3 bar")) + 
geom_line(data=essay_0.3_5_0,aes(x=timestamp,y=pressure_bar,colour="5 cm 0.3 bar")) +

geom_point(data=essay_0.3_5_1,aes(x=timestamp,y=pressure_bar,colour="5 cm 0.3 bar")) + 
geom_line(data=essay_0.3_5_1,aes(x=timestamp,y=pressure_bar,colour="5 cm 0.3 bar")) +

scale_y_continuous(breaks=seq(from = 0, to = 2, by = 0.1)) +
scale_x_continuous(breaks=seq(from = 0, to = 2, by = 0.1)) +
theme_bw(base_size=12) + 
labs(colour="Legend",x="Time (seconds)", y="Pressure(bar)",title="Burst Essay Results in PVA tubes \n (5 mm internal diameter)")+
theme(panel.grid.major.x=element_line(colour="gray90",size=0.3),
     panel.grid.major.y=element_line(colour="gray90",size=0.3),
     panel.grid.minor.x=element_line(colour="gray90",size=0.3),
      panel.grid.minor.y=element_line(colour="gray90",size=0.3))
```
